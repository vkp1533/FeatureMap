```python
import numpy as np
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
tr = pd.read_csv('/content/drive/MyDrive/ML_Datasets/wine_train.csv')
te = pd.read_csv('/content/drive/MyDrive/ML_Datasets/wine_test.csv')
x =  tr.drop('target' , axis = 1)
y = tr['target'].replace(0, -1) # Map 0 to -1 for SVM
print("training data : ",x.shape)
xtest = te.drop('target', axis=1)
ytest = te['target'].replace(0, -1) # Map 0 to -1 for SVM consistency
print("test data size: ", xtest.shape)
```

```python
# parameter initialization
w = np.zeros(x.shape[1])
b= 0
lr = 0.001
lembda = 0.1
itrn = 50
```

```python
# SVM logic  calculating weight and bias
for i in range(itrn):
  for j in range(len(y)):
    cond = y[j] * (np.dot(x.iloc[j], w) + b) >= 1
    if cond:
        w -= lr * (2 * lembda * w)
    else:
        w -= lr * (2 * lembda * w - y[j] * x.iloc[j])
        b += lr * y[j]
print("wieghts =\n",w)
print("bias = ",b)
```

```python

decision = np.dot(xtest, w) + b
mapped = np.where(decision >= 0, 1, -1)
print("Value After calculation: \n ",decision)
print("\nResults after mapping : \n",mapped)
print("\n Actual Target Values\n", ytest.values.reshape(1, -1))
# Calculating accuracy
accuracy = np.mean( mapped == ytest)
print("Test Set Accuracy:",accuracy)
```
