```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
x= np.array ([1,3,5,7,9]).reshape(-1,1)
y= np.array ([1,9,25,49,81])
```

```python
# Now we will map x to polynomial of degree 2 and degree 3
poly2= PolynomialFeatures(degree=2)
poly3= PolynomialFeatures(degree=3)
x_poly2= poly2.fit_transform(x)
x_poly3= poly3.fit_transform(x)

# new feature mapping
mf = pd.DataFrame(x_poly2, columns=['1','x','x^2'])
print(mf)
```

```python
mf2 = pd.DataFrame(x_poly3, columns=['1','x','x^2','x^3'])
print(mf2)
```

```python
from sklearn.linear_model import LinearRegression

# Train polynomial regression models
model2 = LinearRegression()
model2.fit(x_poly2, y)
model3 = LinearRegression()
model3.fit(x_poly3, y)

# Create smooth range of x values for plotting
x_range = np.linspace(1, 9, 100).reshape(-1,1)

# Transform the range using both polynomial mappings
x_range_poly2 = poly2.transform(x_range)
x_range_poly3 = poly3.transform(x_range)

# Predictions for both models
y_pred2 = model2.predict(x_range_poly2)
y_pred3 = model3.predict(x_range_poly3)

# Plot all
plt.figure(figsize=(8,6))
plt.scatter(x, y, color='black', label='Original Data', s=60)
plt.plot(x_range, y_pred2, color='red', label='Degree 2 Fit')
plt.plot(x_range, y_pred3, color='green', label='Degree 3 Fit')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Feature Mapping Visualization (Degree 2 vs Degree 3)')
plt.grid(True)
plt.show()
```
